1. Abra o arquivo "best-selling game consoles.xlsx" em um dataframe.

```python
import pandas as pd
```

```python
game = pd.read_excel("best-selling game consoles.xlsx",
                      sheet_name="consoles")
game.head()
```

2. Substitua a palavra "NES" por "Nitendinho" no nome dos consoles e deixe todos os nomes em mai√∫sculos.

```python
game['Console Name'] = game['Console Name'].str.replace('NES','Nitendinho')
game['Console Name'] = game['Console Name'].str.upper()
game.head()
```

3. Filtre o nome dos consoles com data de release depois de 2010.

```python
game.loc[game['Released Year']>2010,'Console Name']
```

4. De um describe e info da base, substitua os missing values pela string "missing".

```python
game.fillna('missing', inplace = True)
game.info()
```

5. Filtre os consoles que foram descontinuados a menos de 2 anos da data de release.

```python
game['descontinuado'] = game['Discontinuation Year'] - game['Released Year']

game.loc[(game['descontinuado']<=2)&(game['descontinuado']>0),
             ['Console Name',
              'Released Year',
              'Discontinuation Year', 
              'Type', 
              'Company', 
              'Units sold (million)', 
              'Remarks']]
```
